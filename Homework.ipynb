```python
def tinh_tien_dien(chi_so_truoc, chi_so_sau):
    so_kwh = chi_so_sau - chi_so_truoc
    tien = 0
    if so_kwh > 0:
        tieu_thu = min(so_kwh, 50)
        tien += tieu_thu * 1984
        so_kwh -= tieu_thu
    if so_kwh > 0:
        tieu_thu = min(so_kwh, 50)
        tien += tieu_thu * 2050
        so_kwh -= tieu_thu
    if so_kwh > 0:
        tieu_thu = min(so_kwh, 100)
        tien += tieu_thu * 2380
        so_kwh -= tieu_thu
    if so_kwh > 0:
        tieu_thu = min(so_kwh, 100)
        tien += tieu_thu * 2998
        so_kwh -= tieu_thu
    if so_kwh > 0:
        tieu_thu = min(so_kwh, 100)
        tien += tieu_thu * 3350
        so_kwh -= tieu_thu
    if so_kwh > 0:
        tien += so_kwh * 3460
    
    # ThÃªm VAT 8%
    tien = int(round(tien * 1.08))
    return tien
print("Ho va ten: Pham Xuan Toai")
print("Tien phai tra la:", tinh_tien_dien(1020, 1070))

print("\nHo va ten: Nguyen Xuan Long")
print("Tien phai tra la:", tinh_tien_dien(2097, 2278))

print("\nHo va ten: Nguyen Xuan Hai")
print("Tien phai tra la:", tinh_tien_dien(2000, 2586))
```
