```python
#W4A6
a, b = map(int, input().split())
c, d = 0, 0
for t in range(b-a+1):
    a += t
    for i in range(1, a+1):
        if a % i == 0:
            d += 1
    if d == 2:
        c += a
    a -= t
    d = 0 
print(c)
```

```python
#W4A7
a = int(input())
b, c = [],[]
for i in range(1,a+1):
    if a % i == 0:
        b.append(i)
d = 0
for i in b:
    for t in range(1, i+1):
        if i % t == 0:
            d += 1
    if d == 2:
        c.append(i)
    d = 0
print(max(c))
```

```python
#W4A8
a = int(input())
def reverse(i):
    e = 0
    while i > 0:
        e = 10*e + i%10
        i = i//10
    return e
b = {0:a}
c = 0
while b[c] != reverse(b[c]):
    c+=1
    b[c] = b[c-1] + reverse(b[c-1])
print(b[c])
```

```python
#W4A9
import math as m
n = int(input())
b,c = [],[]
for i in range(n+1):
    if m.sqrt(i) == int(m.sqrt(i)):
        b.append(i)
d = 0
for i in b:
    i = str(i)
    for h in range(1,len(i)):
        if all(i[h-1] != i[t] for t in range(h, len(i))):        
            d+=1
    if d == len(i)-1:
        c.append(i)
    d = 0
for i in c:
    print(i, end = ' ')
```

```python
#W4A10
n = int(input())
b = 1
c,e  = {}, {}
for i in range(1, n+1):
    d = i
    while i > 1:
        if i % 2 != 0:
            i = 3*i +1 
            b += 1
        elif i % 2 == 0:
            i = i/2
            b += 1
    c[d] = b
    b = 1
for i in range(1, len(c)+1):
    if c[i] == max(c.values()):
        e[i] = c[i]
for i in range(1,len(c)+1):
    if i == min(e.keys()):
        print(f'({i}, {e[i]})')
```
