```python
a = 7 #Dòng này tạo biến a và gán giá trị là 7
b = 5 #Dòng này tạo biến b và gán giá trị là 5.
c = a - b #Dòng này tính hiệu giữa a và b, sau đó gán kết quả cho biến c
print(c) #Dòng này in giá trị của biến c, tức là 2.
#Kết quả khi chạy mã nguồn trên sẽ in ra:2
```

```python
city = "Hà Nội" #Dòng này tạo một biến city và gán giá trị là chuỗi "Hà Nội", đại diện cho tên thành phố.
year = 2025 #Dòng này tạo một biến year và gán giá trị là 2025, đại diện cho năm.
print("Thành phố: ", city, " - Năm: ", year) #Dòng này in ra một chuỗi, kết hợp tên thành phố và năm với các chuỗi văn bản khác.
#Dòng này in ra một chuỗi, kết hợp tên thành phố và năm với các chuỗi văn bản khác.
#kết quả in ra sẽ là: Thành phố: Hà Nội - Năm: 2025.
```

```python
n = 4        # Gán giá trị 4 cho biến n – số lượng phần tử cần cộng
t = 0        # Khởi tạo biến t với giá trị 0 – dùng để lưu tổng
for i in range(1, n+1):
    t += i #Cộng giá trị i vào biến t (t = t + i)
print(t)  #In ra tổng sau khi đã cộng xong (kết quả là 10)
```

```python
numbers = [1, 2, 3, 4] #Tạo một list chứa 4 phần tử [1, 2, 3, 4].
for x in numbers: #Vòng lặp for sẽ lần lượt lấy từng giá trị trong numbers gán vào biến x.
    if x % 2 == 0:
        print(x, "là số chẵn") #Nếu x chia hết cho 2 (nghĩa là số chẵn) thì điều kiện đúng.Khi đó chạy câu lệnh print(x, "là số chẵn").
    else:
        print(x, "là số lẻ") #Nếu x % 2 != 0 (không chia hết cho 2) thì đây là số lẻ.Chạy câu lệnh print(x, "là số lẻ").
'''Kết quả in ra:
1 là số lẻ
2 là số chẵn
3 là số lẻ
4 là số chẵn'''
```

```python
animals = ["cat", "dog", "cat", "bird"] #Tạo một danh sách animals gồm 4 phần tử: "cat", "dog", "cat", "bird".
count = 0 #Khởi tạo biến đếm count bằng 0.Biến này dùng để đếm số phần tử trong danh sách.
for a in animals: #Vòng lặp for sẽ duyệt qua từng phần tử trong danh sách animals.
    count += 1 #Mỗi lần lặp, count tăng thêm 1.

print("Số phần tử trong danh sách là:", count) #In ra kết quả cuối cùng sau khi đếm.
# Đoạn code này đếm số phần tử trong danh sách bằng cách duyệt qua từng phần tử và tăng biến đếm count.
#Kết quả in ra: Số phần tử trong danh sách là: 4
```
